```python
repo_id = "Mangal-404/phi-2-ResumeRiser-7B"
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer

# Load the model and tokenizer from the Hub
model = AutoModelForCausalLM.from_pretrained(
      repo_id,
      trust_remote_code=True,
      device_map="auto"
)
tokenizer = AutoTokenizer.from_pretrained(repo_id)
```

```python
# input_text = "Which libraries does Mangaleshwaran use for data visualization?"
# input_text = "Where did Mangaleshwaran pursue his Master's degree?"
# input_text = "What score did he obtain in his Bachelor's program?"
# input_text = "What was Mangaleshwaran's academic performance in his Bachelor's degree?"
input_text = "What was the objective of Mangaleshwaran project 'CareerCrafter'?"
```

```python
# Test inference
inputs = tokenizer(input_text, return_tensors="pt").to("cuda")
outputs = model.generate(**inputs, max_length=100)
print(tokenizer.decode(outputs[0], skip_special_tokens=True))
```
